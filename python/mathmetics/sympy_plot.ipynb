```python
# 3d
from sympy import symbols
from sympy.plotting import plot3d

x = [-5, 5]
y = [-5, 5]
z = [-5, 5]
x, y = symbols('x y')
p = plot3d(x*y, (x, -5, 5), (y, -5, 5), show=False)
fig, axe = get_sympy_subplots(p)
axe.plot(xs=x, ys=y, zs=z)
fig.show()
```

```python
# 함수 그리기
class my_func(Function):

    @classmethod
    def eval(cls, x):
        if -1 <= x < 1:
            return -x**2 + 2
        elif 1 <= x <= 3:
            return x**2 - 4*x + 3
        elif 3 < x <= 5:
            return -x**2 + 8*x - 14
        else:
            return None

def my_func2(x):
    return 1/x

my_func3 = lambda x: x**2

plot_function(my_func, is_dot=True)
plot_function(my_func2)
plot_function(my_func3)
plot_function(Heaviside)
```

# Exponential function
$y=a^x$　
- a : base (밑)
- x : exponent (지수)
- $(1/a)^x = a^{-x}$　
- $(a^x)^y = a^{xy}$　
- $\sqrt[x]{a} = a^{\frac{1}{x}}$

```python
plot(2**x, (1/2)**x, -2**x, 2**(1/x), legend=True, **lim(10))
```

# Log
$a^x=b$

$x=log_ab$
- a: base (밑), a>0 & a!=1
- b: antilogarithm (진수), b > 0

- $\log_axy=\log_ax + \log_ay$
- $\log_ax^n=n\log_ax$
- $\log_ab=\log{b}/\log{a} $　
    - 이 특성으로 인해, log를 쓰는 공식에서 밑을 안 쓰는 듯
    - $\log{_{a^n}b}=(\log_ab)/n$
- $b^{\log_a{c}}=c^{\log_a{b}}$

```python
#x, y가 양수여야만 expand_log, logcombine이 적용된다.

x, y = symbols('x y', positive=True)
```

```python
log(x*y), expand_log(log(x*y))
```

```python
log(x/y), expand_log(log(x/y))
```

```python
log(x**2), expand_log(log(x**2))
```

```python
# combine
log(x) + log(y), logcombine(log(x) + log(y))
```

```python
plot(log(x, 2), log(x, 1/2), legend=True, **lim(10))
```

```python
# 지수함수와 log함수는 역함수 관계에 있음
x, y = symbols('x y', positive=True)
inverse_function(2**x, x, y), inverse_function(log(x,2), x, y), is_bijective_function(2**x, x, y), is_bijective_function(log(x,2), x, y)
```

```python
plot(log(x, 2), 2**x, log(x, 1/2), (1/2)**x, legend=True, **lim(10))
```

# 자연로그 & 자연 상수 E
    - 1을 투자해서 1년 후 1의 수익이 나고, 투자 기간만큼 비례해서 수익이 난다고 했을 때,
    - 1을 투자해서 1/2년 후 1/2의 수익이 나고, 1/2년간 재투자해서, 1/2 + 1/4 만큼의 수익을 낼 경우, 1년동안 투자한 것보다 수익이 좋음
    - 이런 식으로, 한정된 기간내에서 투자횟수를 무한히 증가시켰을 때, 특정 값으로 수렴하는 성질이 있고, 이 값을 e로 정의함
    - 극한 개념을 사용할 때, 자연 상수나 로그로 치환되는 경우가 많아서, 사용하는 듯
    - 데이터 과학에서는 점진적 특정 값으로 수렴하는게, 경사 하강등에 도움이 되기 때문에 사용하는 듯.
    - 고등학교 수학에서는 log(a)는 상용로그인데, 대학수학부터는 자연로그가 더 중요하여, log(a)는 자연로그를 의미한다고 함

$\lim_{x \to \infty}(1+\frac{1}{x})^x = e$

$\lim_{x \to 0}(1+x)^{\frac{1}{x}} = e$　
- 극한으로 가기 때문에, x를 어떤 수로 치환 해줘도, 동일한 e값이 됨.
- 여기선 x -> $\frac1 x$로 치환

$\ln{x}$ = $\log_e{x}$　

$\log{x}$ = $\log_{10}{x}$　

$\lim_{x \to 0}\frac{\ln(1+x)}{x} = 1$　

$\lim_{x \to \infty}x\log_a(1 + \frac1 x) = \frac1{\ln{a}}$　

$\lim_{x \to 0}\frac1 x\log_a(1 + x) = \frac1{\ln{a}}$　

$\lim_{x \to 0}\frac{(e^x-1)}{x} = 1$　
- $e^x-1 = t$로 치환
- $x$가 0으로 가면, $t$ 도 0으로 감
- $x=\ln(t+1)$
- $\lim_{t \to 0}\frac{t}{\ln(t+1)} = \lim_{t \to 0}\frac{1}{\frac{1}{t}{\ln(t+1)}} = 1$

$\lim_{x \to 0}\frac{a^x -1}{x} = \ln{a}$　
- $a^x-1 = t$로 치환
- $x$가 0으로 가면, $t$ 도 0으로 감.
- $x=\log_a(t+1), \lim_{t \to 0}\frac{t}{\log_a(t+1)} = \lim_{t \to 0}\frac{1}{\frac{1}{t}{\log_a(t+1)}}$

```python
plot((1+ 1/x)**x, ln(x), legend=True, **lim(10))
```

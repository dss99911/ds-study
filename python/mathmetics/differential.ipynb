# 미분(Differential)

미분 계수 : x=a에서의 순간 변화율(그래프의 특정 지점의 기울기).
## 미분법 공식
- $f^\prime(a) = \lim_{h \to 0}\frac{f(a+h) - f(a)}{h} = \frac{df(x)}{dx}$
-
- $(f(x) \pm g(x))^\prime = f^\prime(x) \pm g^\prime(x)$　
-
- $(cf(x))^\prime = cf^\prime(x)$　
-
- $(f(x)g(x))^\prime = f^\prime(x)g(x) + f(x)g^\prime(x)$　
-
- $\left(\frac{f(x)}{g(x)}\right)^\prime = \frac{f^\prime(x)g(x) - f(x)g^\prime(x)}{g(x)^2}$　


## $f(x) = x^n$의 도함수

- $\frac{d}{dx}(x^n) = nx^{n-1}$　

- $\frac{d}{dx}(c) = 0$　
- $\frac{d}{dx}(1/x) = \frac{n}{x^{n+1}}$　

$ax^2+bx+c=y$에서 $bx+c=y$는 항상 접선
- 두 공식을 빼면, $ax^2=0$ 이 되고, 해는 항상 하나일 수밖에 없으므로
- x=0 에서 접점이 발생($f^\prime(x) = b => 2ax+b = b => x = 0$)

## 삼각함수

### 삼각함수의 극한
$\lim_{x \to \infty}{\frac{\sin{x}}{x}} = 1$

### 삼각함수의 도함수

- $\frac{d}{dx}(\sin{x}) = \cos{x}$　
    - sin(0)일 때, 기울기 1, sin(pi/2)일 때, 기울기 0. 그려보면 일치하는 걸 알 수 있음
- $\frac{d}{dx}(\cos{x}) = -\sin{x}$　
    - cos(0)일 때, 기울기 0, cos(pi/2)일 때, 기울기 -1
- $\frac{d}{dx}(\tan{x}) = \sec^2{x}$　

## 지수함수 로그함수 미분

- $\frac{d}{dx}(e^x) = e^x$　
- $\frac{d}{dx}(a^x) = a^x\ln{a}$　
- $\frac{d}{dx}(\ln{x}) = \frac{1}{x}$　
- $\frac{d}{dx}(\log_a{x}) = \frac{1}{x\ln{a}}$　

## n차 도함수

- $f^{(n)}(x) = \frac{d^ny}{dx^}$　
    - 표현식을 나타낸 것이고, n번 미분하는 방법
- $(f^{-1})^\prime(x)= \frac{1}{f^\prime(f^{-1}(x))}$　
    - 역함수의 도함수. 역함수는 x, y가 바뀌는 형식이므로, 분모로 바뀜

## 합성 함수의 도함수
- $\{f(g(x))\}^\prime = f^\prime(g(x))g^\prime(x)$　

## 음함수의 미분법
- 음함수 정의는 [trigonometric](trigonometric.ipynb) 참조
- $x**2 + y**2 = 1$ 의 미분 => $-x/y$


```python
# x^2 + y^2 = 4 의 미분
idiff(x**2 + y**2 - 4, y, x)
```

```python
# implicit 함수의 접선 그리기
ifun = x ** 2 + y ** 2 - 4
# ifun = x**2 + 2*y**2 - x*y - 3*x - 4*y - 5
p = plot_implicit(ifun, **lim(2), show=False)

fig, ax = get_sympy_subplots(p)

solve(ifun.subs(x, 2), y)
line = idiff_line(ifun, y, x, (sqrt(2), solve(ifun.subs(x, sqrt(2)), y)[0]))
line2 = idiff_line(ifun, y, x, (0, solve(ifun.subs(x, 0), y)[0]))
line3 = idiff_line(ifun, y, x, (1, solve(ifun.subs(x, 1), y)[0]))
plot_function(line, ax=ax, show=False)
plot_function(line2, ax=ax, show=False)
plot_function(line3, ax=ax, show=False)
plt.show()
```

## 로피탈의 정리
- 극한을 구할 때, 부정형(0/0이나 oo/oo)의 극한 값을 구할 때, 보다 쉽게 구하는 방법
- f(x), g(x)가 연속, 미분 가능하고, f(a), g(a)일 때, 0/0이나 oo/oo 의 값이 되는 경우
- 값이 나올 때까지, 여러번 중복해서 적용가능함.

### 사용할 수 없는 경우
- 0/0, oo/oo가 아닌 경우
- 극한값이 없는 진동 값(sin등)이 나오는 경우
- 계속 부정형 값만 나오는 경우.
- 로피탈을 쓰면, 부정형이 더 복잡해지는 경우

$\lim_{x \to a}\frac{f(x)}{g(x)} = \lim_{x \to a}\frac{f^\prime(x)}{g^\prime(x)}$

## 이차 도함수
- 구부러진 정도를 결정
- 이차 도함수가 -라는 것은 기울기가 점차 감소한다는 것, +라는 것은 기울기가 점차 증가한다는 것

극값 : 도함수의 값이 0
    - 극대값 : 도함수의 값이 +에서 -로 바뀌는 점
    - 극소값 : 도함수의 값이 -에서 +로 바뀌는 점

변곡점(inflection point) : 위로 볼록하거나, 아래로 볼록함이 바뀌는 점. 이차 도함수의 값이 0일 때
    - 이차도함수 값이 -이면, 위로 볼록. +이면 아래로 볼록

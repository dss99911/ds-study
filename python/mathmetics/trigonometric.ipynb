# 삼각함수(trigonometric function)

## 원
- 둘레(circumference): $2{\pi}r$
- 넓이 : ${\pi}r^2$
- 시초선 : 기준이 되는 선
- 동경 : 시초선에서 각만큼 기울어진 선
- $x^2 + y^2 = 1$과 같은 형식으로 표현
- 한 x에 대해, y의 값이 두개 있으므로, 함수는 아님.
- 이와 같이, $f(x, y) = 0$ 형태의 관계식이 성립할 때, y를 x의 음함수(implicit function)라고 함. f(x) = y의 경우, y를 x의 양함수(explicit function)

### 호도법(circular measure)
- 각도는 비율이라서, 전체를 무엇으로 할지는 자유이다.
    - 가령 전체를 1로 놓고, 직각을 1/4로 하고, 직각의 호의 길이는 $2{\pi}r/4$ 로 계산해도 됨
- 라디안은 $2{\pi}r$일 때, $2{\pi}$로 표현하여, 둘레와 각의 값이 동일하게 하여, 둘레=각*r 으로 표현이 가능하게 한 방식
    - 각이 $2{\pi}$(rad)일 때, 둘레가 $2{\pi}r$ 임
    - 호의 길이가 반지름과 같을 때 1rad이 된다.
    - 삼각함수에 쓰임

#### radian <-> degree 변환
- $2{\pi}$와 360° 사이의 변환
- $1(rad)=180°/{\pi}$
- $1°={\pi}/180°$


```python
deg(1), rad(1)
```

## 삼각비
- 직각 삼각형에서의 삼각비($x^2+y^2=1$, 반지름이 1인 원 안에 직각삼각형이 있다고 생각)
- $\sin{θ}$ : y/r, r=1
    - (0, 0) -> ($ {\pi}/2$, 1) -> ($\pi $, 0) -> ($3\pi/2 $, -1) -> ($2\pi $, 0)
    - 역수 : $\csc{θ}$ : r/y, cosecant(코시컨트)
- $\cos{θ}$ : x/r, r=1 (co가 together. 붙어있는 쪽)
    - (0, 1) -> ($ {\pi}/2$, 0) -> ($\pi $, -1) -> ($3\pi/2 $, 0) -> ($2\pi $, 1)
    - $\sec{θ}$ : r/x, secant(시컨트)
- $\tan{θ}$ : y/x, tan=> rectangle(사각형의 직각이 있는 쪽)
    - (0, 0) -> ($ {\pi}/2$, 0) -> ($\pi $, -1) -> ($3\pi/2 $, 0) -> ($2\pi $, 1)
    - $\cot{θ}$ : x/y


## 공식
$\sin^2θ+\cos^2θ=1$　
$\tan^2θ+1=\sec^2θ$　
$\cot^2θ+1=\csc^2θ$

$a^2=b^2+c^2-2bc\cos{A}$　
- a,b,c 는 각변, A는 a와 마주보는 각

## 삼각비 값
$\sin\frac{\pi}{6} = \frac{1}{2}$　
$\cos\frac{\pi}{6} = \frac{\sqrt{3}}{2}$　
$\tan\frac{\pi}{6} = \frac{1}{\sqrt{3}}$　

$\sin\frac{\pi}{4} = \frac{1}{\sqrt{2}}$　
$\cos\frac{\pi}{4} = \frac{1}{\sqrt{2}}$　
$\tan\frac{\pi}{4} = 1$　

$\sin\frac{\pi}{3} = \frac{\sqrt{3}}{2}$　
$\cos\frac{\pi}{3} = \frac{1}{2}$　
$\tan\frac{\pi}{3} = \sqrt{3}$　

# 삼각함수의 덧셈정리
$\sin(a+b) = \sin{a}\cos{b} + \cos{a}\sin{b}$　
$\sin(a-b) = \sin{a}\cos{b} - \cos{a}\sin{b}$　
$\cos(a+b) = \cos{a}\cos{b} - \sin{a}\sin{b}$　
$\cos(a-b) = \cos{a}\cos{b} + \sin{a}\sin{b}$　
$\tan(a+b) = \frac{\tan{a}+\tan{b}}{1-\tan{a}\tan{b}}$　
$\tan(a-b) = \frac{\tan{a}-\tan{b}}{1+\tan{a}\tan{b}}$

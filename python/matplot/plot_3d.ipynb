```python
import numpy as np
x = np.linspace(50,150,24)
y = np.linspace(0.5,2.5,24)
x,y = np.meshgrid(x,y)
z = (x-100)**2/y
```

```python
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(projection='3d')

surf = ax.plot3D([-5,5],[-5,5],[-5,5])
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.title('3D')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
ax.plot
surf = ax.plot_surface(x,y,z,rstride=2,cstride=2,cmap=plt.cm.coolwarm,linewidth=0.5,antialiased=False)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.title('3D')
fig.colorbar(surf,shrink=0.5,aspect=5)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
x = np.linspace(50,150,24)
y = np.linspace(0.5,2.5,24)
x,y = np.meshgrid(x,y)
z = (x-100)**2/y

%matplotlib inline

# pyqt를 통해 별도의 창에서 실행 됨. pyqt6는 안되고, 5만 지원
# pip install PyQt5
# jupyter껐다가 이것 먼저 실행해야 작동 하는 듯
%matplotlib qt


fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x,y,z,rstride=2,cstride=2,cmap=plt.cm.coolwarm,linewidth=0.5,antialiased=False)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.title('3D')
fig.colorbar(surf,shrink=0.5,aspect=5)
```
